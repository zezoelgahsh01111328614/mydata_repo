```python
def value_to_class(value):
    clas=0
    if value>=400000:
         clas=1
    elif value>=300000:
         clas=2
    elif value>=200000:
         clas=3
    elif value>=100000:
         clas=4
    else:
         clas=5
    return clas

    
```

```python
df['median_house_class']=df['median_house_value'].apply(value_to_class)
```
