```python
import pandas as pd
from glob import glob
```

# using glob method to select all files from a bath

```python
all_monthes = sorted(glob(r'C:\Users\Options\Desktop\SalesAnalysis\Sales_Data\Sales_*_*.csv'))
all_monthes
```

# concat all files in one variable 

```python
all_months_data =pd.concat((pd.read_csv(file).assign(filename = file)
        for file in all_monthes) ,ignore_index = True)
```
